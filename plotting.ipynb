```python
# Plot potential 1D
import numpy as np
import matplotlib.pyplot as plt
import functions
from functions import *
plt.title('Potential')
plt.xlabel(r"$\frac{x}{r}$")
plt.ylabel(r"$\frac{V}{\hbar\omega}$")
for i in range (1,6):
    L_over_r = 3
    N = i*20
    epsilon = L_over_r/N
    D = 1
    
    epsilon_2 = epsilon**2
    functions.N = N
    functions.D = D
    functions.mu = 10
    functions.epsilon_2 = epsilon_2
    x = np.zeros((N,)*D)
    psi = np.ones_like(x)
    potential = potential_function(psi)
    x_axis = np.linspace(-N/2,N/2,N)
    x_axis= x_axis*epsilon
    


    plt.plot(x_axis, potential, label = fr'N={N}, $\epsilon$={epsilon:.1e}')


plt.legend(loc='upper right')

plt.show()
```

```python
# Plot potential 2D
plt.title('Potential, 2D')
plt.xlabel(r"$\frac{x}{r}$")
plt.ylabel(r"$\frac{x}{r}$")

L_over_r = 2
N = 100
epsilon = L_over_r/N
D = 2
mu = 10
epsilon_2 = epsilon**2
functions.N = N
functions.D = D
functions.mu = mu
functions.epsilon_2 = epsilon_2
x = np.zeros((N,)*D)
psi = np.ones_like(x)
potential = potential_function(psi)
x_axis = np.linspace(-N/2,N/2,N)
x_axis= x_axis*epsilon
plt.imshow(potential.real, extent=[-N/2,N/2, -N/2,N/2])
plt.text(N/10, N/10, fr'N={N}, $\epsilon$={epsilon:.1e}', bbox={'facecolor': 'white', 'pad': 10})
plt.colorbar().set_label(r"$\frac{V}{\hbar\omega}$")

plt.show()
```
