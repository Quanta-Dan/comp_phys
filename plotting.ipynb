```python
# Plot potential 1D
import numpy as np
import matplotlib.pyplot as plt
import functions
from functions import *
plt.title('Potential')
plt.xlabel(r"$\frac{x}{r}$")
plt.ylabel(r"$\frac{V}{\hbar\omega}$")
for i in range (1,6):
    L_over_r = 3
    N = i*20
    epsilon = L_over_r/N
    D = 1
    
    epsilon_2 = epsilon**2
    functions.N = N
    functions.D = D
    functions.mu = 10
    functions.epsilon_2 = epsilon_2
    x = np.zeros((N,)*D)
    psi = np.ones_like(x)
    potential = potential_function(psi)
    x_axis = np.linspace(-N/2,N/2,N)
    x_axis= x_axis*epsilon
    


    plt.plot(x_axis, potential, label = fr'N={N}, $\epsilon$={epsilon:.1e}')


plt.legend(loc='upper right')

plt.show()
```

```python
# Plot potential 2D
plt.title('Potential, 2D')
plt.xlabel(r"$\frac{x}{r}$")
plt.ylabel(r"$\frac{x}{r}$")

L_over_r = 2
N = 100
epsilon = L_over_r/N
D = 2
mu = 10
epsilon_2 = epsilon**2
functions.N = N
functions.D = D
functions.mu = mu
functions.epsilon_2 = epsilon_2
x = np.zeros((N,)*D)
psi = np.ones_like(x)
potential = potential_function(psi)
x_axis = np.linspace(-N/2,N/2,N)
x_axis= x_axis*epsilon
plt.imshow(potential.real, extent=[-N/2,N/2, -N/2,N/2], cmap = 'plasma')
plt.text(-N/2.5,N/2.5, fr'N={N}, $\epsilon$={epsilon:.1e}', bbox={'facecolor': 'white', 'pad': 10})
plt.colorbar().set_label(r"$\frac{V}{\hbar\omega}$")

plt.show()
```

```python
#plot plane wave 2D

import numpy as np
import matplotlib.pyplot as plt
import functions
from functions import *
L_over_r = 3
N = 200
epsilon = L_over_r/N
D = 2
mu = 10
epsilon_2 = epsilon**2
functions.N = N
functions.D = D
functions.mu = mu
functions.epsilon_2 = epsilon_2
y = generate_plane_wave((3,1))



x_axis = np.linspace(-N/2,N/2,N)
y_axis = np.linspace(-N/2,N/2,N)
X,Y = np.meshgrid(x_axis*epsilon,y_axis*epsilon)


hf = plt.figure(figsize=(10,10))
ha = hf.add_subplot(111, projection='3d')
ha.view_init(elev=30, azim=30)
ha.set_proj_type('ortho')
plt.title(fr'Plane Wave for N={N}, $\epsilon^{2}$={epsilon_2:.1e}')
plt.xlabel(r"$\frac{x}{r}$")
plt.ylabel(r"$\frac{x}{r}$")
#plt.zlabel(r"$\Psi$")

ha.set_zlabel(r"$\Psi$")
#plt.plot(x_axis,lowest_EigVector, label = fr'Lowest eigenvector for N={N}, $\epsilon$={epsilon_2:.1e}')

ha.plot_surface(X, Y, y.real, cmap='plasma', rstride=1, cstride=1, linewidth=0, alpha=0.9)

plt.show()


```
